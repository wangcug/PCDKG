### loading the NER and tokenizer model.

```python
nlp = stanza.Pipeline(lang='en', processors='tokenize,ner',                        
   ner_model_path='/home/wcb/Desktop/stanza-train7/stanza/saved_models/ner/en_por_nertagger.pt')
#tokenize_model_path='/home/wcb/Desktop/stanza-train7/stanza/saved_models/tokenize/en_por_tokenizer.pt',
```

### Input the test data 

```python
f = open('/home/wcb/Desktop/test.txt','r', encoding='utf-8')
testdata=f.read()
f.close()
```

```python
test = nlp(testdata)
```

```python
for i, sentence in enumerate(test.sentences):
    print(f'====== Sentence {i+1} tokens =======')
    print(*[f'id: {token.id}\ttext: {token.text}' for token in sentence.tokens], sep='\n')
```

```python
# Output the entity recognizaiton results
print(*[f'entity: {ent.text}\ttype: {ent.type}' for sent in test.sentences for ent in sent.ents], sep='\n')
```
